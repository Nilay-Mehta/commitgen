```python
from datasets import load_dataset
ds = load_dataset("Maxscha/commitbench", split="train", streaming=True)
sample = list(ds.take(100))
print(f"Pulled {len(sample)} rows")
```

```python
first = sample[0]
print("Field names:", list(first.keys()))
for k, v in first.items():
    vtype = type(v).__name__
    preview = str(v)[:80].replace("\n", "\\n")
    print(f"  {k}: ({vtype}) {preview}")

# Surface actual language label format (NOT a stats histogram -
# we just need to see "Python" vs "python" vs "py" etc.)
lang_field = "lang" if "lang" in first else ("language" if "language" in first else None)
if lang_field:
    unique_langs = set(str(r.get(lang_field)) for r in sample)
    print(f"\nUnique values in `{lang_field}` (sample of 100): {unique_langs}")
else:
    print("\nNo lang/language field found - check schema above.")
```

```python
import random
random.seed(42)
for ex in random.sample(sample, 5):
    print("=" * 80)
    print("MESSAGE:")
    print(ex.get("message") or ex.get("msg") or "<no message field>")
    print("\nDIFF (first 1000 chars):")
    diff_text = ex.get("diff") or ex.get("patch") or "<no diff field>"
    print(str(diff_text)[:1000])
```
